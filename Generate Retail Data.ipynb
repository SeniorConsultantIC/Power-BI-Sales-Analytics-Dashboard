```python
"""
CSV FILE GENERATOR FOR POWER BI STAR SCHEMA
Generates 6 CSV files with 500 rows each
Run this script to create all data files
"""

import pandas as pd
import random
from datetime import datetime, timedelta
import os

# Set random seed for reproducibility
random.seed(42)

def create_all_csv_files():
    """Generate all 6 CSV files with 500 rows each"""
    
    print("🚀 Generating Star Schema Dataset...")
    print("=" * 50)
    
    # ==================== 1. STORE TABLE (Dimension) ====================
    print("\n📊 Generating Store Table (500 rows)...")
    stores = []
    store_types = ['Flagship', 'Standard', 'Compact', 'Express', 'Superstore']
    regions = ['North', 'South', 'East', 'West', 'Central', 'Northeast', 'Southeast', 'Midwest']
    cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 
              'San Antonio', 'San Diego', 'Dallas', 'Austin', 'Boston', 'Seattle', 
              'Denver', 'Miami', 'Atlanta', 'Detroit', 'Portland', 'Las Vegas']
    
    for i in range(1, 501):
        store_id = f"S{i:03d}"
        store_name = f"Store_{random.choice(['Plaza', 'Mall', 'Center', 'Market', 'Hub'])} {i}"
        region = random.choice(regions)
        city = random.choice(cities)
        store_type = random.choice(store_types)
        opening_date = datetime(2015, 1, 1) + timedelta(days=random.randint(0, 3000))
        
        stores.append({
            'StoreID': store_id,
            'StoreName': store_name,
            'Region': region,
            'City': city,
            'StoreType': store_type,
            'OpeningDate': opening_date.strftime('%Y-%m-%d')
        })
    
    df_store = pd.DataFrame(stores)
    df_store.to_csv('Store Table.csv', index=False)
    print(f"   ✓ Created: Store Table.csv ({len(df_store):,} rows)")
    
    # ==================== 2. PRODUCT TABLE (Dimension) ====================
    print("\n📦 Generating Product Table (500 rows)...")
    products = []
    categories = ['Electronics', 'Home', 'Furniture', 'Clothing', 'Sports', 'Beauty', 'Toys', 'Automotive']
    subcategories = {
        'Electronics': ['Computers', 'Phones', 'Accessories', 'TVs', 'Audio', 'Cameras'],
        'Home': ['Kitchen', 'Appliances', 'Decor', 'Garden', 'Tools', 'Lighting'],
        'Furniture': ['Office', 'Living Room', 'Bedroom', 'Storage', 'Outdoor', 'Kids'],
        'Clothing': ['Men', 'Women', 'Kids', 'Accessories', 'Shoes', 'Sports Wear'],
        'Sports': ['Exercise', 'Outdoor', 'Team Sports', 'Fitness', 'Camping', 'Cycling'],
        'Beauty': ['Skincare', 'Makeup', 'Hair', 'Fragrance', 'Tools', 'Bath'],
        'Toys': ['Educational', 'Action Figures', 'Games', 'Outdoor Toys', 'Dolls', 'Puzzles'],
        'Automotive': ['Parts', 'Tools', 'Accessories', 'Interior', 'Exterior', 'Fluids']
    }
    
    for i in range(1, 501):
        product_id = f"P{i:03d}"
        category = random.choice(categories)
        subcategory = random.choice(subcategories[category])
        product_name = f"{category}_{subcategory}_Product_{i}"
        unit_price = round(random.uniform(9.99, 1999.99), 2)
        product_cost = round(unit_price * random.uniform(0.4, 0.7), 2)
        
        products.append({
            'ProductID': product_id,
            'ProductName': product_name,
            'Category': category,
            'SubCategory': subcategory,
            'UnitPrice': unit_price,
            'ProductCost': product_cost
        })
    
    df_product = pd.DataFrame(products)
    df_product.to_csv('Product Table.csv', index=False)
    print(f"   ✓ Created: Product Table.csv ({len(df_product):,} rows)")
    
    # ==================== 3. CUSTOMER TABLE (Dimension) ====================
    print("\n👥 Generating Customer Table (500 rows)...")
    customers = []
    age_groups = ['18-24', '25-34', '35-44', '45-54', '55+']
    membership_tiers = ['Bronze', 'Silver', 'Gold', 'Platinum']
    
    first_names = ['John', 'Sarah', 'Mike', 'Emily', 'David', 'Lisa', 'Robert', 'Maria', 'James', 'Patricia',
                   'Michael', 'Jennifer', 'William', 'Linda', 'Richard', 'Susan', 'Joseph', 'Jessica', 'Thomas', 'Karen',
                   'Charles', 'Nancy', 'Christopher', 'Betty', 'Daniel', 'Margaret', 'Matthew', 'Sandra', 'Anthony', 'Ashley']
    last_names = ['Smith', 'Johnson', 'Brown', 'Davis', 'Wilson', 'Anderson', 'Taylor', 'Garcia', 'Martinez', 'Lee',
                  'White', 'Harris', 'Martin', 'Thompson', 'Robinson', 'Clark', 'Lewis', 'Walker', 'Hall', 'Allen']
    
    for i in range(1, 501):
        customer_id = f"C{i:03d}"
        customer_name = f"{random.choice(first_names)} {random.choice(last_names)}"
        age_group = random.choice(age_groups)
        membership_tier = random.choice(membership_tiers)
        city = random.choice(cities)
        signup_date = datetime(2020, 1, 1) + timedelta(days=random.randint(0, 1600))
        
        customers.append({
            'CustomerID': customer_id,
            'CustomerName': customer_name,
            'AgeGroup': age_group,
            'MembershipTier': membership_tier,
            'City': city,
            'SignupDate': signup_date.strftime('%Y-%m-%d')
        })
    
    df_customer = pd.DataFrame(customers)
    df_customer.to_csv('Customer Table.csv', index=False)
    print(f"   ✓ Created: Customer Table.csv ({len(df_customer):,} rows)")
    
    # ==================== 4. FACT SALES TABLE ====================
    print("\n💰 Generating Fact Sales (500 rows)...")
    sales = []
    start_date = datetime(2023, 1, 1)
    end_date = datetime(2024, 12, 31)
    
    for i in range(1, 501):
        sale_id = f"SL{i:04d}"
        date_key = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
        store_id = f"S{random.randint(1, 500):03d}"
        product_id = f"P{random.randint(1, 500):03d}"
        customer_id = f"C{random.randint(1, 500):03d}"
        quantity = random.randint(1, 10)
        unit_price = round(random.uniform(9.99, 1999.99), 2)
        sales_amount = round(quantity * unit_price, 2)
        
        sales.append({
            'SaleID': sale_id,
            'DateKey': date_key.strftime('%Y-%m-%d'),
            'StoreID': store_id,
            'ProductID': product_id,
            'CustomerID': customer_id,
            'Quantity': quantity,
            'SalesAmount': sales_amount
        })
    
    df_sales = pd.DataFrame(sales)
    df_sales.to_csv('Fact_Sales.csv', index=False)
    print(f"   ✓ Created: Fact_Sales.csv ({len(df_sales):,} rows)")
    
    # ==================== 5. FACT RETURNS TABLE ====================
    print("\n🔄 Generating Fact Returns (500 rows)...")
    returns = []
    return_reasons = ['Defective', 'Wrong Item', 'Not Satisfied', 'Damaged', 'Better Price Elsewhere', 'No Longer Needed']
    
    for i in range(1, 501):
        return_id = f"RT{i:04d}"
        date_key = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
        store_id = f"S{random.randint(1, 500):03d}"
        product_id = f"P{random.randint(1, 500):03d}"
        customer_id = f"C{random.randint(1, 500):03d}"
        return_reason = random.choice(return_reasons)
        quantity = random.randint(1, 3)
        refund_amount = round(quantity * random.uniform(9.99, 1999.99), 2)
        
        returns.append({
            'ReturnID': return_id,
            'DateKey': date_key.strftime('%Y-%m-%d'),
            'StoreID': store_id,
            'ProductID': product_id,
            'CustomerID': customer_id,
            'ReturnReason': return_reason,
            'Quantity': quantity,
            'RefundAmount': refund_amount
        })
    
    df_returns = pd.DataFrame(returns)
    df_returns.to_csv('Fact_Returns.csv', index=False)
    print(f"   ✓ Created: Fact_Returns.csv ({len(df_returns):,} rows)")
    
    # ==================== 6. FACT TARGET TABLE ====================
    print("\n🎯 Generating Fact Target (500 rows)...")
    targets = []
    years = [2023, 2024, 2025]
    
    for i in range(1, 501):
        target_id = f"TG{i:04d}"
        year = random.choice(years)
        month = random.randint(1, 12)
        store_id = f"S{random.randint(1, 500):03d}"
        product_id = f"P{random.randint(1, 500):03d}"
        product_price = random.uniform(9.99, 1999.99)
        sales_target = round(product_price * random.randint(50, 500), 2)
        quantity_target = random.randint(50, 500)
        
        targets.append({
            'TargetID': target_id,
            'Year': year,
            'Month': month,
            'StoreID': store_id,
            'ProductID': product_id,
            'SalesTarget': sales_target,
            'QuantityTarget': quantity_target
        })
    
    df_target = pd.DataFrame(targets)
    df_target.to_csv('Fact_Target.csv', index=False)
    print(f"   ✓ Created: Fact_Target.csv ({len(df_target):,} rows)")
    
    # ==================== SUMMARY ====================
    print("\n" + "=" * 50)
    print("✅ ALL 6 CSV FILES GENERATED SUCCESSFULLY!")
    print("=" * 50)
    print("\n📁 Files created:")
    print("   1. Store Table.csv (Dimension - 500 rows)")
    print("   2. Product Table.csv (Dimension - 500 rows)")
    print("   3. Customer Table.csv (Dimension - 500 rows)")
    print("   4. Fact_Sales.csv (Fact Table - 500 rows)")
    print("   5. Fact_Returns.csv (Fact Table - 500 rows)")
    print("   6. Fact_Target.csv (Fact Table - 500 rows)")
    
    # Show sample of each file
    print("\n" + "=" * 50)
    print("📊 SAMPLE DATA PREVIEW:")
    print("=" * 50)
    
    print("\n🔹 Store Table (first 3 rows):")
    print(df_store.head(3).to_string(index=False))
    
    print("\n🔹 Product Table (first 3 rows):")
    print(df_product.head(3).to_string(index=False))
    
    print("\n🔹 Customer Table (first 3 rows):")
    print(df_customer.head(3).to_string(index=False))
    
    print("\n🔹 Fact Sales (first 3 rows):")
    print(df_sales.head(3).to_string(index=False))
    
    print("\n🔹 Fact Returns (first 3 rows):")
    print(df_returns.head(3).to_string(index=False))
    
    print("\n🔹 Fact Target (first 3 rows):")
    print(df_target.head(3).to_string(index=False))
    
    print("\n" + "=" * 50)
    print("🎉 READY TO IMPORT INTO POWER BI!")
    print("=" * 50)

if __name__ == "__main__":
    create_all_csv_files()
```
